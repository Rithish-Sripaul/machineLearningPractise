```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris = iris[['petal_length', 'petal_width']]
dataset_heading = ["petal_length", "petal_width"]
x_train = iris.iloc[:, 0].values
y_train = iris.iloc[:, 1].values
x_train = x_train.reshape(-1, 1)
```

```python
def compute_train_model(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    
    for i in range(m):
        f_wb[i] = w*x[i] + b
    
    return f_wb
```

```python
sgdr = SGDRegressor(max_iter=3000)
sgdr.fit(x_train, y_train)
print(sgdr)
```

```python
b = sgdr.intercept_
w = sgdr.coef_
print(f"Coefficeint: {w}")
print(f"Intercept: {b}")
```

```python
y_pred = sgdr.predict(x_train)
print(y_pred[:10])
print(y_train[:10])
```

```python
f_wb = compute_train_model(x_train, w, b)
plt.scatter(x_train, y_train, marker="x", c='r')
plt.plot(x_train, f_wb, c="b")
plt.show()
```
